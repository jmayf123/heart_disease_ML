```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/cleaned_data.csv', index_col=0)
```

```python
df['Heart Disease'] = np.where(df['Heart_Disease'] == 1, 'Yes', 'No')
df['Sex'] = np.where(df['Sex'] == 1, 'Male', 'Female')
```

```python
df['Heart Disease'].value_counts()
```

```python
hd = list((df['Heart_Disease'].value_counts()))
plt.pie(hd, labels=['No Heart Disease', 'Heart Disease'], autopct='%1.1f%%')
plt.show()
```

Need to work on Class imbalance correction with resampling. 

```python
sns.countplot(data = df, x = 'Sex', hue = 'Heart Disease')
plt.ylabel('Count')
```

```python
sns.histplot(data=df, x='Age_Category', hue='Sex', multiple='dodge', shrink=2.8)
plt.xticks(np.arange(0, 15, step=1))
plt.show()
```

Age Distributions for those with HD and without HD

```python
df_hd = df[df['Heart_Disease'] == 1]
df_no_hd = df[df['Heart_Disease'] == 0]
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,5))

sns.histplot(ax=axes[0],x=df_no_hd['Age_Category'])
sns.histplot(ax=axes[1],x=df_hd["Age_Category"])
axes[0].set_title("No - Heart disease")
axes[1].set_title("Yes - Heart Disease")
axes[0].set_xticks((np.arange(0, 15, step=1)))
axes[1].set_xticks((np.arange(0, 15, step=1)))

plt.show()
```
